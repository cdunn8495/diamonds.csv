```python
import pandas as pd

url = 'https://raw.githubusercontent.com/cdunn8495/diamonds.csv/main/diamonds.csv'

df = pd.read_csv(url, index_col=0)

print("First seven rows:")
print(df.head(7))


print("\n Last seven rows:")
print(df.tail(7))

numerical_stats = df[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']]
print("\n Descriptive Statistics for Numerical Columns:")
print(numerical_stats)

categorical_stats = df[['cut', 'color', 'clarity']].describe()
print("Descriptive Statistics for Categorical Columns:")
print(categorical_stats)

unique_cut_values = df['cut'].unique()
unique_color_values = df['color'].unique()
unique_clarity_values = df['clarity'].unique()
print("Unique Cut Values:", unique_cut_values)
print("Unique Color Values:", unique_color_values)
print("Unique Clarity Values:", unique_clarity_values)

%matplotlib inline

df[['carat', 'depth', 'price', 'table', 'x', 'y', 'z']].hist(figsize=(12, 10))
```
